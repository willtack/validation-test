Copy heatmaps to stage folders and rename according to lookup table

```python
import os
import pandas as pd
import shutil

datadir = '../heatmaps/stage2' # where the unaliased files are
stage3dir = '../heatmaps/stage3'
stage4dir = '../heatmaps/stage4'
df = pd.read_csv('../lookup_table_2021-12-1_17:25.csv')

for hmap in os.listdir(datadir):    
    # read lookup table for subject pseudonyms for stage 2
    sub_label = hmap.split('sub-')[1].split('_')[0] # get subject label from filename
    three_n = int(df[df['Subject Label']==sub_label]['Stage 3'])
    four_n = int(df[df['Subject Label']==sub_label]['Stage 4'])
    hmap_path = os.path.join(datadir, hmap)
    print(f"{sub_label} : {three_n} : {four_n} : {hmap_path}")
    
    #  create filename with subject alias
    stage3path = os.path.join(stage3dir, f"Subject{three_n}.png")
    stage4path = os.path.join(stage4dir, f"Subject{four_n}.png")
    # copy file
    shutil.copyfile(hmap_path, stage3path)
    shutil.copyfile(hmap_path, stage4path)
    print(f"Copying to {stage3path}")
    print(f"Copying to {stage4path}")
```
